```python
tickers = ['A', 'AAL', 'AAP', 'AAPL', 'ABBV', 'ABC', 'ABT']
data = yf.download(tickers, start='2015-01-01', end='2022-03-25')

data['Adj Close'].plot(subplots=True,figsize=(12,12))
```

```python
#Compute the daily returns of each stock and plot a histogram of the returns:

returns = data['Adj Close'].pct_change()
returns.dropna(inplace=True)
returns.hist(bins=50, figsize=(12, 8))
```

```python
#Compute the correlation matrix of the stocks and plot it as a heatmap:

import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix of the stocks
corr_matrix = data['Adj Close'].corr()

# Create a heatmap of the correlation matrix
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=.5, ax=ax)
ax.set_title('Correlation Matrix of S&P 500 Stocks')
plt.show()
```

```python
#Compute the rolling mean and standard deviation of each stock and plot them on a single graph:

window_size = 60
rolling_mean = data['Adj Close'].rolling(window_size).mean()
rolling_std = data['Adj Close'].rolling(window_size).std()

fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(data['Adj Close'], label='Adj Close')
for ticker in tickers:
    ax.plot(rolling_mean[ticker], label=f'{ticker} rolling mean')
    ax.plot(rolling_std[ticker], label=f'{ticker} rolling std')
ax.legend(loc='upper left')
```
